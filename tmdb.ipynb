```python
import requests

def get_data(API_key, year, genre_id, sort_value, pageno):
    base_url = "https://api.themoviedb.org/3"
    url = (
        f"{base_url}/discover/movie?"
        f"api_key={API_key}"
        f"&primary_release_year.gte={year}"
        f"&with_genres={genre_id}"
        f"&sort_by={sort_value}"
        f"&page={pageno}"
    )

    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        return data['results']
    else:
        return {'error': response.status_code, 'message': response.text}
```

```python
get_data(API_key,2000,35,'populairty.desc',1)
```

```python
import pandas as pd
df=pd.DataFrame(columns=['Movie ID','Movie Name'])
pageno = 1
while(len(df)<300):
    query_results = get_data(API_key,2000,35,'popularity.desc',pageno)
    if query_results=='error':
        continue
    else:
        for movie in query_results:
            df.loc[len(df)] = [movie['id'],movie['title']]
        pageno+=1
```

```python
def get_similar_movie(API_key,movie_id:str):
    query_url = base_url + '/movie/'+ str(movie_id)+'/similar?api_key='+API_key
    response = requests.get(query_url)
    if response.status_code == 200:
       data = response.json()
       ids = []
       for movie in data['results']:
           ids.append(movie['id'])
       return ids
    else:
        return 'error'
```
